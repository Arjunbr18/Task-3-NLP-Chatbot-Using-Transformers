## **Introduction**
A chatbot is a software application that interacts with users using natural language. With the advancement of Artificial Intelligence (AI) and Natural Language Processing (NLP), chatbots have become more intelligent and capable of understanding human conversations.

In this project, we build a chatbot using a pre-trained transformer model (FLAN-T5) from Hugging Face. The chatbot can take user input, process it, and generate meaningful and accurate responses dynamically.

## **Objective**
To build a conversational chatbot using a pre-trained transformer model that can interact with users and generate meaningful and accurate responses.

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
```

```python
# Loading FLAN-T5 model (better for correct answers)

model_name = "google/flan-t5-base"

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print("Model loaded successfully!")
```

```python
print("\nChatbot: Hello! I am your AI assistant. How can I help you today?")
print("Type 'exit' or 'quit' to stop the chatbot.")

while True:
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() in ['exit', 'quit']:
        print("Chatbot: Goodbye! Have a nice day.")
        break

    text = user_input.lower()

    # Controlled responses for important queries
    if "artificial intelligence" in text or "what is ai" in text:
        print("Chatbot: Artificial Intelligence is the simulation of human intelligence in machines that can learn, reason, and make decisions.")
        continue

    elif "who created python" in text or ("python" in text and "who" in text):
        print("Chatbot: Python was created by Guido van Rossum and released in 1991.")
        continue

    elif "hello" in text or "hi" in text:
        print("Chatbot: Hello! How can I assist you today?")
        continue

    elif "thank" in text:
        print("Chatbot: You're welcome! Feel free to ask more questions.")
        continue

    # Transformer fallback (for other questions)
    prompt = f"Answer clearly and correctly: {user_input}"

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    outputs = model.generate(
        **inputs,
        max_length=100,
        do_sample=False
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    print("Chatbot:", response.strip())
```

## **Pipeline Flow**

User Input   **→**   Tokenization   **→**   Model Processing   **→**   Response Generation   **→**   Display Output   **→**   Loop Until Exit

## **Explanation of the Code**

- **Model Loading**
  - A pre-trained transformer model (FLAN-T5) is loaded from Hugging Face using the Transformers library.
  - FLAN-T5 is an instruction-tuned model that generates more accurate and meaningful responses compared to regular conversational models.

- **User Input**
  - The chatbot takes input from the user through the console using the `input()` function.

- **Response Generation**
  - The model generates responses based on the input using the `generate()` function.
  - An instruction-based prompt is used to guide the model to produce clear and correct answers.

- **Conversation Flow**
  - The chatbot runs in a loop, allowing continuous interaction with the user.

- **Exit Condition**
  - The chatbot stops when the user enters "exit" or "quit".
